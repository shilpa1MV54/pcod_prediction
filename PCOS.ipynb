data standardization

```python


# Assuming 'Follicle No (L)' is the problematic column
# 1. Replace 'a' with NaN:
X['Follicle No (L)'] = pd.to_numeric(X['Follicle No (L)'], errors='coerce')

# 2. Impute NaN values with the mean of the column (or another strategy):
X['Follicle No (L)'] = X['Follicle No (L)'].fillna(X['Follicle No (L)'].mean())

# Ensure all columns are numeric before applying StandardScaler:
# Convert all columns to numeric, errors='coerce' will convert invalid values to NaN
for column in X.columns:
    X[column] = pd.to_numeric(X[column], errors='coerce')

# Now impute NaN for all columns
X = X.fillna(X.mean())

# 3. Apply StandardScaler:
scaler = StandardScaler()
scaler.fit(X)
```

```python
standardized_data=scaler.transform(X)
```

```python
X=standardized_data
Y=pcodi_data['PCOS (Y/N)_x']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
classifier=svm.SVC(kernel='linear')
```

```python
print(Y.dtype)
```

```python

from sklearn.preprocessing import LabelEncoder
# Assuming 'Y' is your original target variable

# Convert the target variable 'Y' to numerical labels using LabelEncoder
label_encoder = LabelEncoder()
Y_encoded = label_encoder.fit_transform(Y)

# Now, use the encoded target variable for train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_encoded, test_size=0.2, stratify=Y_encoded, random_state=2)

# Fit the classifier using the encoded target variable
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('accuracy on training data:',training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('accuracy on training data:',test_data_accuracy)
```
